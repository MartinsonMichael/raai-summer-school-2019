```python
from matplotlib import pyplot as plt
import numpy as np
from gym_road_cars import CarRacing, DummyCar
import cv2
import os

import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import matplotlib.animation as animation
```

```python
env = CarRacing()
```

```python
state_pic, _ = env.reset(2)
```

```python
plt.figure(figsize=(16, 12))
plt.imshow(state_pic[:1000, :1000], origin='lower')

# wheels
plt.scatter(
    env._agent_car.get_wheels_positions()[:, 0], 
    env._agent_car.get_wheels_positions()[:, 1], 
    color='blue'
)
# wheels vectors
plt.scatter(
    env._agent_car.get_wheels_positions()[:, 0] + 30 * env._agent_car.get_wheels_vectors[:, 0], 
    env._agent_car.get_wheels_positions()[:, 1] + 30 * env._agent_car.get_wheels_vectors[:, 1], 
    color='magenta'
)

plt.scatter(env._agent_car.hull.position.x, env._agent_car.hull.position.y, color='green')
plt.scatter(
    env._agent_car.hull.position.x + 30 * env._agent_car.get_car_vector[0], 
    env._agent_car.hull.position.y + 30 * env._agent_car.get_car_vector[1], 
    color='magenta'
)
for wheel in env._agent_car.iterate_over_front_wheels():
    plt.scatter(wheel.position.x, wheel.position.y, color='red')
plt.show()
```
